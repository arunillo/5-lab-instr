```python
# Установка необходимых библиотек
!pip install squarify

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from scipy import stats
import squarify

# Настройка отображения графиков
plt.rcParams['figure.figsize'] = (12, 8)
plt.rcParams['font.size'] = 12

print("Лабораторная работа №5. Вариант 20")
print("=" * 50)

# 1. Загрузка данных
print("\n1. Загрузка данных Yeast Data Set...")

url = "https://archive.ics.uci.edu/ml/machine-learning-databases/yeast/yeast.data"
columns = ['sequence_name', 'mcg', 'gvh', 'alm', 'mit', 'erl', 'pox', 'vac', 'nuc', 'class']
try:
    df = pd.read_csv(url, delim_whitespace=True, header=None, names=columns)
    print("Данные успешно загружены")
except Exception as e:
    print(f"Ошибка загрузки: {e}")
    # Альтернативная ссылка
    url = "https://raw.githubusercontent.com/jbrownlee/Datasets/master/yeast.csv"
    df = pd.read_csv(url, header=None, names=columns)

print(f"Размерность данных: {df.shape}")
print(f"Столбцы: {list(df.columns)}")

# Выделяем нужные признаки и класс
features = df[['mcg', 'gvh', 'alm']].copy()
class_labels = df['class'].copy()

print("\nПервые 5 строк признаков:")
print(features.head())
print("\nСтатистика признаков:")
print(features.describe())
print("\nРаспределение классов:")
print(class_labels.value_counts())

# 2. Обработка пропущенных значений и выбросов
print("\n" + "="*50)
print("2. Обработка пропущенных значений и выбросов...")

# Проверка пропущенных значений в метках классов
missing_class = class_labels.isnull().sum()
print(f"Пропущенных значений в метках классов: {missing_class}")

if missing_class > 0:
    mask = class_labels.notnull()
    features = features[mask]
    class_labels = class_labels[mask]
    print(f"Удалено записей с пропущенными метками: {missing_class}")

# Проверка пропущенных значений в признаках
missing_features = features.isnull().sum().sum()
print(f"Общее количество пропущенных значений в признаках: {missing_features}")

if missing_features > 0:
    missing_percentage = (features.isnull().sum().sum() / len(features)) * 100
    print(f"Процент записей с пропущенными значениями: {missing_percentage:.2f}%")

    # Заменяем пропущенные значения медианой класса
    print("Замена пропущенных значений медианой класса...")
    for feature in features.columns:
        for class_name in class_labels.unique():
            mask = (class_labels == class_name) & (features[feature].isnull())
            if mask.sum() > 0:
                class_median = features[class_labels == class_name][feature].median()
                features.loc[mask, feature] = class_median
                print(f"Заменено {mask.sum()} пропусков в признаке {feature} для класса {class_name} на значение {class_median:.3f}")
else:
    print("Пропущенных значений в признаках не обнаружено")
    # Удаляем выбросы с помощью z-score
    print("Удаление выбросов с помощью z-score...")
    z_scores = np.abs(stats.zscore(features))
    outliers_mask = (z_scores > 3).any(axis=1)
    outliers_count = outliers_mask.sum()
    outliers_percentage = (outliers_count / len(features)) * 100

    print(f"Обнаружено выбросов: {outliers_count}")
    print(f"Процент выбросов: {outliers_percentage:.2f}%")

    if outliers_percentage < 5:
        # Увеличиваем порог для удаления минимум 5% точек
        threshold = np.percentile(z_scores.max(axis=1), 95)
        outliers_mask = (z_scores.max(axis=1) > threshold)
        outliers_count = outliers_mask.sum()
        outliers_percentage = (outliers_count / len(features)) * 100
        print(f"После корректировки порога:")
        print(f"Удалено выбросов: {outliers_count}")
        print(f"Процент удаленных точек: {outliers_percentage:.2f}%")

    features = features[~outliers_mask]
    class_labels = class_labels[~outliers_mask]

print(f"Итоговый размер данных: {len(features)} записей")

# 3. Гистограмма для дискретизированного признака с наибольшей дисперсией
print("\n" + "="*50)
print("3. Построение гистограммы...")

variances = features.var()
max_var_feature = variances.idxmax()
max_variance = variances[max_var_feature]

print(f"Признак с наибольшей дисперсией: '{max_var_feature}' (дисперсия: {max_variance:.4f})")

# Дискретизируем признак на 10 бинов
features_discrete = pd.cut(features[max_var_feature], bins=10, include_lowest=True)

# Создаем сводную таблицу для stacked histogram
cross_tab = pd.crosstab(features_discrete, class_labels)

plt.figure(figsize=(14, 8))
ax = cross_tab.plot(kind='bar', stacked=True, color=plt.cm.Set3(np.linspace(0, 1, len(cross_tab.columns))))
plt.title(f'Гистограмма частот для дискретизированного признака {max_var_feature}', fontsize=16, pad=20)
plt.xlabel(f'Интервалы признака {max_var_feature}', fontsize=12)
plt.ylabel('Количество записей', fontsize=12)
plt.legend(title='Класс', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.xticks(rotation=45, ha='right')
plt.grid(axis='y', alpha=0.3)
plt.tight_layout()
plt.show()

# 4. Area Chart для среднего значения признака с наибольшей дисперсией по классам
print("\n" + "="*50)
print("4. Построение Area Chart...")

class_means = features.groupby(class_labels)[max_var_feature].mean().sort_values(ascending=False)
max_class = class_means.idxmax()
max_value = class_means[max_class]

print(f"Класс с наибольшим средним значением: '{max_class}' (значение: {max_value:.3f})")

plt.figure(figsize=(14, 8))

# Area chart
plt.fill_between(range(len(class_means)), class_means.values, alpha=0.6, color='skyblue', label=f'Среднее {max_var_feature}')
plt.plot(range(len(class_means)), class_means.values, 'o-', color='navy', alpha=0.8)

plt.title(f'Area Chart: Зависимость среднего значения {max_var_feature} от класса', fontsize=16, pad=20)
plt.xlabel('Классы', fontsize=12)
plt.ylabel(f'Среднее значение {max_var_feature}', fontsize=12)
plt.xticks(range(len(class_means)), class_means.index, rotation=45)

# Аннотация для класса с наибольшим средним значением
max_index = list(class_means.index).index(max_class)
plt.annotate(f'Максимум\n{max_class}\n({max_value:.3f})',
             xy=(max_index, max_value),
             xytext=(max_index, max_value + 0.1),
             ha='center',
             bbox=dict(boxstyle='round,pad=0.3', facecolor='yellow', alpha=0.7),
             arrowprops=dict(arrowstyle='->', connectionstyle='arc3,rad=0.1', color='red'))

plt.grid(True, alpha=0.3)
plt.tight_layout()
plt.show()

# 5. Treemap для распределения точек по классам
print("\n" + "="*50)
print("5. Построение Treemap...")

class_counts = class_labels.value_counts()
print("Распределение по классам:")
for class_name, count in class_counts.items():
    percentage = (count / len(class_labels)) * 100
    print(f"  {class_name}: {count} записей ({percentage:.1f}%)")

plt.figure(figsize=(16, 10))
colors = plt.cm.Set3(np.linspace(0, 1, len(class_counts)))

# Создаем Treemap
squarify.plot(sizes=class_counts.values,
              label=[f'{label}\n{count} ({count/len(class_labels)*100:.1f}%)'
                    for label, count in class_counts.items()],
              color=colors,
              alpha=0.7,
              text_kwargs={'fontsize': 10})

plt.title('Treemap: Распределение точек данных по классам', fontsize=16, pad=20)
plt.axis('off')

# Создаем легенду
legend_elements = [plt.Rectangle((0,0),1,1, facecolor=color, label=f'{label} ({count})')
                  for label, count, color in zip(class_counts.index, class_counts.values, colors)]
plt.legend(handles=legend_elements, title='Классы', bbox_to_anchor=(1, 1), loc='upper left')

plt.tight_layout()
plt.show()

# 6. Пузырьковая диаграмма
print("\n" + "="*50)
print("6. Построение пузырьковой диаграммы...")

# Находим два признака с наибольшей дисперсией
sorted_variances = variances.sort_values(ascending=False)
top_two_features = sorted_variances.index[:2].tolist()
remaining_feature = sorted_variances.index[2]

print(f"Два признака с наибольшей дисперсией: {top_two_features}")
print(f"Признак для размера пузырьков: '{remaining_feature}'")

# Нормализуем размер пузырьков
bubble_sizes = (features[remaining_feature] - features[remaining_feature].min()) / \
               (features[remaining_feature].max() - features[remaining_feature].min()) * 200 + 20

plt.figure(figsize=(16, 12))

# Создаем scatter plot с разными цветами и маркерами для классов
unique_classes = class_labels.unique()
colors = plt.cm.tab10(np.linspace(0, 1, len(unique_classes)))
markers = ['o', 's', '^', 'D', 'v', '<', '>', 'p', '*', 'h', 'X', 'P']

for i, class_name in enumerate(unique_classes):
    mask = class_labels == class_name
    plt.scatter(features.loc[mask, top_two_features[0]],
                features.loc[mask, top_two_features[1]],
                s=bubble_sizes[mask],
                c=[colors[i]],
                marker=markers[i % len(markers)],
                alpha=0.7,
                label=class_name,
                edgecolors='black',
                linewidth=0.5)

# Выделяем область между медианой и максимумом для обоих признаков
x_median = features[top_two_features[0]].median()
x_max = features[top_two_features[0]].max()
y_median = features[top_two_features[1]].median()
y_max = features[top_two_features[1]].max()

plt.axvspan(x_median, x_max, ymin=(y_median - features[top_two_features[1]].min()) /
            (features[top_two_features[1]].max() - features[top_two_features[1]].min()),
            ymax=(y_max - features[top_two_features[1]].min()) /
            (features[top_two_features[1]].max() - features[top_two_features[1]].min()),
            alpha=0.3, color='red', label='Область: медиана-максимум')

plt.xlabel(top_two_features[0], fontsize=14)
plt.ylabel(top_two_features[1], fontsize=14)
plt.title(f'Пузырьковая диаграмма: {top_two_features[0]} vs {top_two_features[1]}\n'
          f'(Размер пузырьков зависит от {remaining_feature})', fontsize=16, pad=20)

plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left')
plt.grid(True, alpha=0.3)

# Добавляем информацию о выделенной области
plt.text(0.05, 0.95, f'Выделенная область:\n'
                     f'{top_two_features[0]}: [{x_median:.2f}, {x_max:.2f}]\n'
                     f'{top_two_features[1]}: [{y_median:.2f}, {y_max:.2f}]',
         transform=plt.gca().transAxes, bbox=dict(boxstyle='round', facecolor='wheat', alpha=0.8),
         verticalalignment='top')

plt.tight_layout()
plt.show()

print("\n" + "="*50)
print("ОТЧЕТ ЗАВЕРШЕН")
print(f"Обработано записей: {len(features)}")
print(f"Использованные признаки: {list(features.columns)}")
print(f"Количество классов: {len(class_labels.unique())}")

# Сохранение результатов
print("\nСохранение отчета...")
from google.colab import files

# Сохраняем обработанные данные
processed_df = features.copy()
processed_df['class'] = class_labels
processed_df.to_csv('yeast_processed.csv', index=False)

print("Файл 'yeast_processed.csv' готов для скачивания")
files.download('yeast_processed.csv')
```
